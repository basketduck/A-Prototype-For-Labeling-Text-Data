```python
import csv
import smart_open
import gensim
from collections import defaultdict
from gensim import corpora

with smart_open.open('/Users/hiphone/Downloads/ads-100,000.csv') as f:
    ads_csv = csv.reader(f, delimiter = ',')
    headers = next(ads_csv)
    documents = list(map(lambda x: x[1], ads_csv))
```

```python
model = gensim.models.doc2vec.Doc2Vec.load('/Users/hiphone/fb_10w_d100_e10.d2v')
```

```python
selected_document_id = 30858
print(list(map(lambda x: x[0], model.docvecs.most_similar(selected_document_id, topn = 30))))
```

```python
similar_docs = list(map(lambda x: x[0], model.docvecs.most_similar(selected_document_id, topn = 30)))
print(documents[selected_document_id])
print('---------')
for i in similar_docs:
    print(documents[i])
    print('---------')
```
