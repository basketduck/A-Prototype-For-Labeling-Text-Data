```python
import csv
import smart_open
import gensim
from collections import defaultdict
from gensim import corpora
# with open('/Users/hiphone/Downloads/ads_all.csv') as f:
with smart_open.open('/Users/hiphone/Downloads/ads-100,000.csv') as f:
    ads_csv = csv.reader(f, delimiter = ',')
    headers = next(ads_csv)
    documents = list(map(lambda x: x[1], ads_csv))
#     documents = [line[1] for line in ads_csv]
```

```python
train_corpus = []
test_corpus = []
for i, document in enumerate(documents):
    tokens = gensim.utils.simple_preprocess(document) 
    if i%10 == 0:
        test_corpus.append(tokens)
    else:
        train_corpus.append(gensim.models.doc2vec.TaggedDocument(tokens, [i]))
print(train_corpus[0:2])
print(test_corpus[0:2])
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=100, min_count=2, epochs=10, workers = 4)
```

```python
model.build_vocab(train_corpus)
```

```python
model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vector = model.infer_vector(test_corpus[3])
print(vector)
```

```python
model.save('./fb_10w_d100_e10.d2v')
```
